```python
import os
import pathlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import models
from IPython import display
from scipy.io.wavfile import write

# Set the seed value for experiment reproducibility.
seed = 42
tf.random.set_seed(seed)
np.random.seed(seed)

from google.colab import drive
drive.mount('/content/drive')

DATASET_PATH = "/content/drive/My Drive/dataset/luganda_data_set"
data_dir = pathlib.Path(DATASET_PATH)

commands = np.array(tf.io.gfile.listdir(str(data_dir)))
commands = commands[(commands != 'README.md') & (commands != '.DS_Store')]
print('Commands:', commands)

train_ds, val_ds = tf.keras.utils.audio_dataset_from_directory(
    directory=data_dir,
    batch_size=64,
    validation_split=0.2,
    seed=0,
    output_sequence_length=16000,
    subset='both')

label_names = np.array(train_ds.class_names)
print()
print("label names:", label_names)

def write_wav(audio, labels):
    def write_and_print_info(audio):
        wav_file = 'output.wav'
        sample_rate = 16000
        np_audio = np.array(audio).astype(np.int16)
        write(wav_file, sample_rate, np_audio)
        print(f"WAV file: {wav_file}")
        print(f"Sample rate: {sample_rate}")
        print(f"Number of bytes per sample: {np_audio.itemsize}")

    audio = tf.py_function(write_and_print_info, [audio], tf.int64)
    return audio, labels

train_ds = train_ds.map(write_wav, tf.data.AUTOTUNE)
val_ds = val_ds.map(write_wav, tf.data.AUTOTUNE)

test_ds = val_ds.shard(num_shards=2, index=0)
val_ds = val_ds.shard(num_shards=2, index=1)

# Training loop
for example_audio, example_labels in train_ds.take(1):
    # This will trigger the write_wav function
    print("Training Example:")
    print(example_audio.shape)
    print(example_labels.shape)

# Validation loop
for example_audio, example_labels in val_ds.take(1):
    # This will trigger the write_wav function
    print("\nValidation Example:")
    print(example_audio.shape)
    print(example_labels.shape)
```
